First of all, we import some usefull libraries.

```python
from typing import List
import nltk as nltk
```

We load Trump's speech.

```python
file = open("./datasets/trump.txt", "r",encoding="utf-8") 
trump = file.read()
```

NLTK allows to get words from sentences. We're dividing a text into separate words with _word_tokenizer_ method from NLTK.

```python
words = nltk.word_tokenize(trump)
print("Tokens: " + str(words))
```

 Then, we get tags with _pos tag_ method from NLTK.

```python
tags = nltk.pos_tag(words)
print("Tagged: " + str(tags))
```
